# Git 에서 Github까지


1. Git 

- **작업 directory(폴더)** : 현재 작업 중인 파일이 위치한 디렉토리(폴더), git 이 추적하는 파일과 수정 내역이 저장 

- **Staging Area** : git 저장소에 올라갈 파일이 임시로 저장되는 공간

- **로컬 Repository** : 전역 (로컬) git 으로 관리되는 저장소 .git/ 이라는 저장소(DB)에 저장됨 >> 코드에 대한 버전 관리를 가능하게 함 (이력 추적, 롤백 가능)

- **원격 Repository** : github 서버에 저장된 원격 저장소로 협업 공간 (pull / push / clone)

### add, commit

```python
#terminal 에서 실행 

#git init : 폴더/파일 생성 
git init .git #작업 디렉토리 폴더 생성 
git init guide.ipynb #guide.ipynb 파일 생성 
```

```python
#git 사용자 등록 
git config user.name "Your Name"
git config user.email "Your Email"@example.com

#사용자의 전체 환경에서 설정 
git config --global user.name "Your Name"
git config --global user.email "Your Email"@example.com

#사용자 등록 해제 
#특정 작업 폴더에만 설정된 사용자 정보 삭제
git config --unset user.name
git config --unset user.email

#전체 환경에서 설정된 사용자 정보 삭제
git config --global --unset user.name
git config --global --unset user.email
```

```python
#git status : 현재 상태 확인 (현재 인식하고 있는 프로젝트 관련 내용들 출력)
git status
```

```python
#1) git add : 스테이징 영역에 파일 추가(수정사항이 있는 파일을 스테이징 영역에 올림)
git add 파일명 #특정 파일 추가
git add . #현재 디렉토리의 모든 파일 추가
git add *.ipynb #확장자가 ipynb인 모든 파일 추가

#2) git commit : 스테이징 영역에 있는 파일을 로컬 저장소에 커밋
git commit -m "첫 번째 커밋"
#-m : 커밋 메시지 작성 옵션
git commit -am "두 번째 커밋:메세지" #파일 내용 변경시/ 새로운 파일 생성시 x 
#-a : 수정된 파일을 자동으로 스테이징 영역에 추가 후 커밋
#(단, 새로 생성된 파일은 제외)
```

```python
#git log : 커밋 기록 확인(commit 된 내역들 출력, 조회 프로젝트의 커밋 내역, 작성자, 날짜, 메세지 확인 )
git log
git log --oneline #한 줄로 요약된 커밋 기록 확인
git log --graph #그래프로 시각화된 커밋 기록 확인   

#git show : 특정 커밋의 상세 내역 확인
git show 커밋ID
#커밋ID : git log 명령어로 확인한 커밋 해시값 (앞의 몇 글자만 입력해도 무방)
```

```python
#git rm : 삭제/ 작업 디렉토리에서 파일 삭제 및 삭제 내용 스테이징 영역에 추가(git의 추적 대상에서도 제거))
git rm 파일명 #파일 삭제 및 스테이징 영역에 추가
git rm --cached 파일명 #파일 삭제는 하지 않고 스테이징 영역에서만 제거
git rm -r 폴더명/ #폴더 및 하위 파일들 삭제 및 스테이징 영역에 추가 
#-r : 폴더 및 하위 파일들 삭제 옵션

#cf. git rm -> git commit 으로 커밋 해야 삭제 내용이 로컬 저장소에 반영됨, 복구 불가

#1) git restore : 수정/삭제된 파일을 마지막 커밋 상태로 복원

#작업 디렉토리에서 파일 복원(물리적으로 삭제되었을 때는 여기에서만 삭제됨)
git restore 파일명 

#--staged : 스테이징 영역에서 복원 옵션. (삭제되었던 기록 남음)
git restore --staged 파일명 #스테이징 영역에서 파일 복원

#============================================================
#2) git reset : 커밋 기록 및 스테이징 상태를 이전 상태로 되돌리는 명령어
#특정 커밋 상태로 복원(스테이징 영역 및 작업 디렉토리 모두 복원)
git reset 커밋ID 

#특정 커밋 상태로 복원(스테이징 영역은 복원 되지 않음)
#--soft : 스테이징 영역은 복원되지 않는 옵션                            
git reset --soft 커밋ID 

#(스테이징 영역 및 작업 디렉토리 모두 복원, 복원 이후의 기록 모두 삭제)
#--hard : 복원 이후의 기록 모두 삭제 옵션   
git reset --hard 커밋ID #특정 커밋 상태로 복원


#특정 커밋 상태로 복원(스테이징 영역만 복원, 작업 디렉토리는 복원되지 않음)
git reset --mixed 커밋ID 
```

### Branch

- 독립적인 작업 흐름 생성하기 위한 Git 의 기능 , 메인에서 분리된 별도의 개발 경로 

- 하나의 프로젝트에 대해 여러 사람이 간섭 없이 독자적인 기능 개발 가능 

ex) 

master/main --> branch prod // branch dev

```python
#branch 목록 출력 및 생성 
git branch #현재 작업에 대한 브랜치 목록 출력 
git branch 브랜치명 #새로운 브랜치 생성 

#작업 브랜치 바꾸기 
git switch 브랜치명 #특정 브랜치로 작업 디렉토리 전환
git switch -c 브랜치명 #새로운 브랜치를 생성하고 해당 브랜치로 전환
#* 브랜치 생성 시점에는 파일 구조가 동일하다 

#작업 합치기 
git merge 브랜치명(dev/prod) #현재 작업 중인 브랜치에 특정 브랜치의 작업 내용 병합

#conflict(충돌) 가 발생했다면? 
#동일한 파일의 같은 부분이 두 브랜치에서 서로 다르게 수정된 경우 
#git 이 결정을 내릴 수 없어 작업자가 수동으로 충돌난 부분을 해결해주어얗한다. 
1)충돌난 파일을 열어 충돌된 부분을 수정한 후 
2)git add 파일명 
3)git commit -m "충돌 해결 커밋 메세지" 

#작업 합치기 취소 
#병합 도중 충돌 문제가 발생했을 때 병합을 중단하고 병합 시도 전 상태로 되돌리기
git merge --abort

#branch 삭제
git branch -d 브랜치명 #로컬 브랜치 삭제
git branch -D 브랜치명 #강제 로컬 브랜치 삭제
```

### 원격저장소 작업관리

```python
#원격 저장소 연결 : 로컬 repo <--> 원격 repo 연결 
#git remote 
git remote add origin <원격 저장소 URL> #원격 저장소 추가

#원격 저장소 확인 : 로컬 repo에 연결된 원격 repo 확인 
git remote -v 
#-v : 원격 저장소의 URL 정보까지 함께 출력 옵션

#원격 저장소 연결 해제 
git remote rm origin #(연결시 1:1 대응 연결 )
```

- github -> local

```python
#원격 저장소 가져오기 (github -> local)
#원격 Repo의 모든 파일, 커밋 이력, 브랜치 등을 로컬 Repo로 가져오는
git clone <원격 저장소 URL>
# 원격 repo >> 로컬 repo, 작업 디렉토리 
```

```python

#원격 저장소 파일 가져오기 
#원격 저장소의 변경 사항을 로컬 저장소로 병합해 다운로드하여 동기화

#1) github에서 브랜치 생성 

#2) 원격저장소에서 신규 브랜치 정보 가져오기 
git fetch origin #원격 저장소의 변경 사항을 로컬 저장소로 가져오기  

#3) 가져온 브랜치를 로컬 저장소에 생성
git branch 브랜치명 origin/브랜치명 

#4) 가져온 브랜치를 작업 디렉토리에 반영
git pull origin 브랜치명 #가져올 디렉토리/브랜치 이름 경로 설정 후 
```

- local -> github

```python
#local -> github
#원격 저장소에 파일 저장 : 로컬 저장소의 변경 사항을 원격 저장소로 업로드하여 동기화
git push origin 브랜치명

#원격 저장소에 브랜치를 생성하여 현재 작업중인 파일들을 
#해당 브랜치에 저장하는 명령어 
#깃헙에서 새로운 브랜치 생성도 가능 
git push -u origin [브랜치 명]
#origin : 원격 저장소 이름
#브랜치명 : 원격 저장소에 반영할 로컬 브랜치    
```

```python
#1) 깃헙 비어있는 저장소에 기존 로컬 저장소 연결하기
git remote add origin <원격 저장소 URL>

#2) 로컬에서 파일 생성 및 커밋 후 push 
git init 파일명
git add 파일명
git commit -m "메세지"
git push -u origin main #(또는 master)  
```

### Pull Request merge의 세가지 옵션

Create a merge commit

- 두 브랜치의 변경 사항을 모두 유지하면서 병합되며 merge commit 생성된다.
각 브랜치의 변경 사항이 과거의 커밋으로 보존되고, 새로운 커밋이 추가되어 최종 병합이 완료 됨.
✅ → 브랜치의 히스토리를 모두 유지하며 merge가 가능하여 빠르고 정확하게 진행 상황 추적 가능
❗ → 협업자가 많을수록 commit log가 빠르게 복잡해질 수 있다.

Squash and Merge

-  브랜치에서의 모든 변경 사항을 하나의 커밋으로 압축하여 병합하는 방식
각각의 커밋에서 발생한 모든 변경 사항을 병합 후에 하나의 새로운 커밋을 생성
✅ → 중요한 커밋 히스토리만을 남김으로써 핵심 내용들만 이력에 담을 수 있다.
❗ → 각 커밋에 대한 개별적인 맥락, 작업자 정보 등 상세 이력이 하나로 합쳐져 사라짐.

Rebase and Merge

-    특정 branch의 merge 상황을 master(main) 브랜치 위에 재위치(rebase)시킨 후 병합하는 방식.
모든 커밋이 개별적으로 유지되면서, 마치 처음부터 main에서 작업한 것처럼 보이게 됨.
✅ → 깨끗하고 선형적인 커밋 히스토리를 만들어 주어 히스토리 파악이 용이
❗ → 관련 커밋 ID들이 바뀌어 혼란스러울 수 있음.
